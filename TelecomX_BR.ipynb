#📌 Extracão

```python
import pandas as pd
```

```python
url = pd.read_json('https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json')
```

```python
dados = url
dados.head()
```

#🔧 Transformação

```python
dados.info()
```

```python
colunas_df = list(dados.columns)
colunas_df
```

```python
dados = pd.json_normalize(
    dados[['customerID', 'Churn', 'customer', 'phone', 'internet', 'account']].to_dict('records'),
    sep='_'
)
display(dados.head())
```

```python
colunas_df = list(dados.columns)
colunas_df
```

```python
dados.head()
```

```python
dados.set_index('customerID', inplace=True)
dados.head()
```

```python
dados.info()
```

```python
import numpy as np

dados['account_Charges_Total'] = (
    dados['account_Charges_Total']
    .replace(' ', np.nan)  # Replace space with NaN
    .astype(str)
    .str.replace(',', '.', regex=False)
    .str.replace('R\$', '', regex=True)
    .str.strip()
)

dados['account_Charges_Total'] = pd.to_numeric(dados['account_Charges_Total'], errors='coerce')
```

```python
dados.info()
```

```python
#Verificar nulos
dados.isnull().sum()
```

```python
#Visualizar linhas com valores nulos
dados[dados.isnull().any(axis=1)]
```

```python
#Preencher valores nulos em account_Charges_Total
mask = dados['account_Charges_Total'].isnull()

dados.loc[mask, 'account_Charges_Total'] = (
    dados.loc[mask, 'account_Charges_Monthly'] *
    dados.loc[mask, 'customer_tenure']
)
```

```python
#Checar quantidade de valores nulos após o preenchimento
print("\nValores nulos após o preenchimento:")
print(dados['account_Charges_Total'].isnull().sum())
```

```python
#Visualizar valores nulos preenchidos
print("\nExemplo de linhas preenchidas:")
print(dados.loc[mask])
```

```python
#Conferir presença de duplicados
dados.duplicated().sum()
```

Foi conferido que todos os ID do cliente são diferentes, logo esses valores são falsos duplicados.

```python
dados[dados.duplicated()]
```

```python
#Conferir categorias inconsistentes
for col in dados.select_dtypes(include='object').columns:
    print(f"\n{col} → {dados[col].unique()}")
```

```python
print(dados['Churn'].value_counts())
```

```python
dados = dados[dados['Churn'].str.strip() != ""]
print(dados['Churn'].value_counts())
```

```python
# Criar um dicionário de mapeamento para converter "Yes" e "No" em binários
map_binario = {"Yes": 1, "No": 0}

# Aplicar a transformação nas colunas desejadas
colunas_binarias = ['Churn', 'customer_Partner', 'customer_Dependents', 'phone_PhoneService', 'account_PaperlessBilling']

for col in colunas_binarias:
    dados[col] = dados[col].map(map_binario)
```

```python
#Exibir colunas binárias
dados[colunas_binarias].head()
```

```python
#Padronizar e traduzir os nomes das colunas
dados = dados.rename(columns={
    'customer_gender': 'genero_cliente',
    'customer_SeniorCitizen': 'maior_65_anos',
    'customer_Partner': 'possui_parceiro',
    'customer_Dependents': 'dependentes',
    'customer_tenure': 'tempo_contrato_meses',
    'account_Charges_Monthly': 'cobranca_mensal',
    'account_Charges_Total': 'cobranca_total',
    'account_Contract': 'tipo_contrato',
    'phone_PhoneService': 'servico_telefonico',
    'phone_MultipleLines': 'mais_de_uma_linha_telefonica',
    'internet_InternetService': 'servico_internet',
    'internet_OnlineSecurity': 'servico_seguranca_online',
    'internet_OnlineBackup': 'servico_online_backup',
    'internet_DeviceProtection': 'servico_protecao_dispositivo',
    'internet_TechSupport': 'servico_suporte_tecnico',
    'internet_StreamingTV': 'servico_streaming_TV',
    'internet_StreamingMovies': 'servico_streaming_filmes',
    'account_PaperlessBilling': 'fatura_online',
    'account_PaymentMethod': 'forma_de_pagamento'

})
dados.head()
```

```python
# Criar coluna Contas_Diarias
dados['conta_diaria'] = dados['cobranca_mensal'] / 30
dados[['cobranca_mensal', 'conta_diaria']].head()
```

Contas_Diarias fornece uma visão mais detalhada do comportamento financeiro diário do cliente.

Essa coluna é útil para análises de tendência, comparação entre clientes e cálculos de receita diária agregada.

# 📄 Relatório Final de Análise de Churn de Clientes

## 1. Introdução

O objetivo deste relatório é analisar os dados de clientes da empresa de telecomunicações Telecom X, identificando os principais fatores que levam à evasão de clientes, conhecida como Churn.

Compreender esses fatores é essencial para desenvolver estratégias eficazes de retenção, reduzindo assim perdas de receita. A análise se concentra em explorar as características dos clientes que cancelam seus serviços em comparação com aqueles que permanecem ativos.

Nas primeiras etapas do Challenge, o foco foi compreender o funcionamento do dataset disponibilizado pela Alura/Oracle. Para isso, foram realizados os seguintes passos:


### 2. Extração e Preparação dos Dados

**a. Fonte dos dados:** Arquivo TelecomX_Data.json, disponibilizado pela Alura/Oracle.

**b. Local de armazenamento**: data/dadosAPI/TelecomX_Data.json.

**c. Processos realizados:**
* Extração do dataset via API.
* Inspeção inicial para identificação de tipos de dados, valores nulos e inconsistentes.
* Normalização de dados compactados com pd.json_normalize().
* Padronização de valores "Yes"/"No" para 1 e 0.
* Tradução e renomeação de colunas para português.
* Criação da coluna conta_diaria para análise de gastos diários.
* Remoção de registros com valores em branco na coluna Churn.
* Tratamento de valores nulos na coluna account_Charges_Total usando a regra: cobrança mensal × tempo de contrato (meses) para contratos anuais.

### 3. Limpeza e Validação

**Verificação de duplicatas:** 22 registros duplicados foram identificados, porém esses registros possuem ID do cliente diferentes, logo não foram considerados verdadeiros duplicados.

**Valores ausentes:** 11 valores nulos foram tratados de acordo com a regra de negócio -> para clientes com contrato anual, o valor total foi calculado multiplicando a cobrança mensal pelo tempo de contrato em meses. Após esse procedimento, verificou-se que não havia mais valores nulos.

**Valores inconsistentes:** corrigidos e padronizados para facilitar análises futuras.

OBS: Observei a presença de 224 registros que têm valores em branco na coluna Churn. Visto que esse valor representa apenas 3% da base, e essa coluna é muito importante para análise de churn por ser a variável-alvo, optei por excluí-los.

## Conclusões e Insights

Com base na análise exploratória, podemos destacar os seguintes insights:

*   **Taxa de Churn:** Aproximadamente 26.5% dos clientes cancelaram seus serviços, representando uma taxa de evasão significativa.
*   **Tipo de Contrato:** Clientes com contratos mensais apresentam uma probabilidade muito maior de churn em comparação com aqueles com contratos de um ou dois anos. Isso sugere que contratos de longo prazo promovem maior retenção.
*   **Tempo de Contrato (Tenure):** Clientes com menor tempo de contrato (principalmente nos primeiros 12 meses) têm uma taxa de churn consideravelmente mais alta. A taxa de evasão diminui à medida que o tempo de contrato aumenta.
*   **Idosos:** Clientes com mais de 65 anos demonstram uma taxa de churn maior do que clientes mais jovens.
*   **Dependentes:** Clientes sem dependentes tendem a ter uma taxa de churn maior do que aqueles com dependentes.
*   **Método de Pagamento:** Clientes que utilizam o método de pagamento "Electronic check" parecem ter uma taxa de churn mais alta em comparação com outros métodos.
*   **Serviços Contratados:** Há uma tendência de que clientes com menos serviços contratados apresentem maior churn. Clientes com 0 ou 1 serviço têm uma taxa de evasão notavelmente alta. Clientes com um maior número de serviços tendem a permanecer mais tempo.
*   **Gastos:** Clientes que cancelam tendem a ter uma cobrança mensal mais alta e um gasto total acumulado menor, o que faz sentido dado o menor tempo de contrato.

## Recomendações

Com base nas conclusões, sugiro as seguintes recomendações para reduzir a evasão de clientes:

*   **Programas de Fidelidade para Contratos Longos:** Oferecer incentivos e benefícios exclusivos para clientes que optam por contratos de um ou dois anos.
*   **Estratégias de Retenção para Novos Clientes:** Implementar ações de acompanhamento proativo e suporte dedicado para clientes nos primeiros meses de contrato, que são mais propensos ao churn.
*   **Ofertas Específicas para Idosos:** Desenvolver planos e serviços adaptados às necessidades de clientes com mais de 65 anos, com foco em simplicidade e suporte acessível.
*   **Pacotes de Serviços:** Promover a contratação de múltiplos serviços, destacando os benefícios e a economia de pacotes combinados, para aumentar o número de serviços por cliente.
*   **Análise do Método de Pagamento "Electronic Check":** Investigar as razões pelas quais clientes que usam este método têm maior churn. Pode haver problemas com o processo de pagamento, taxas ou insatisfação geral.
*   **Segmentação e Campanhas Direcionadas:** Utilizar os insights sobre os grupos de maior risco (contratos mensais, baixo tenure, idosos, sem dependentes, poucos serviços) para criar campanhas de marketing e retenção personalizadas.

Este relatório fornece uma base sólida para entender o churn de clientes. Análises mais aprofundadas, como a construção de modelos preditivos de churn, podem ser realizadas para identificar individualmente os clientes com maior probabilidade de evasão e implementar ações de retenção direcionadas.
