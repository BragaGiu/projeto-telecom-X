```python
import pandas as pd
```

```python
url = pd.read_json('https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json')
```

```python
dados = url
dados.head()
```

#ðŸ”§ TransformaÃ§Ã£o

```python
dados.info()
```

```python
colunas_df = list(dados.columns)
colunas_df
```

```python
dados = pd.json_normalize(
    dados[['customerID', 'Churn', 'customer', 'phone', 'internet', 'account']].to_dict('records'),
    sep='_'
)
display(dados.head())
```

```python
colunas_df = list(dados.columns)
colunas_df
```

```python
dados.head()
```

```python
dados.set_index('customerID', inplace=True)
dados.head()
```

```python
dados.info()
```

```python
import numpy as np

dados['account_Charges_Total'] = (
    dados['account_Charges_Total']
    .replace(' ', np.nan)  # Replace space with NaN
    .astype(str)
    .str.replace(',', '.', regex=False)
    .str.replace('R\$', '', regex=True)
    .str.strip()
)

dados['account_Charges_Total'] = pd.to_numeric(dados['account_Charges_Total'], errors='coerce')
```

```python
dados.info()
```

```python
#Verificar nulos
dados.isnull().sum()
```

```python
#Visualizar linhas com valores nulos
dados[dados.isnull().any(axis=1)]
```

```python
#Preencher valores nulos em account_Charges_Total
mask = dados['account_Charges_Total'].isnull()

dados.loc[mask, 'account_Charges_Total'] = (
    dados.loc[mask, 'account_Charges_Monthly'] *
    dados.loc[mask, 'customer_tenure']
)
```

```python
#Checar quantidade de valores nulos apÃ³s o preenchimento
print("\nValores nulos apÃ³s o preenchimento:")
print(dados['account_Charges_Total'].isnull().sum())
```

```python
#Visualizar valores nulos preenchidos
print("\nExemplo de linhas preenchidas:")
print(dados.loc[mask])
```

```python
#Conferir categorias inconsistentes
for col in dados.select_dtypes(include='object').columns:
    print(f"\n{col} â†’ {dados[col].unique()}")
```

```python
print(dados['Churn'].value_counts())
```

```python
dados = dados[dados['Churn'].str.strip() != ""]
print(dados['Churn'].value_counts())
```

```python
# Criar um dicionÃ¡rio de mapeamento para converter "Yes" e "No" em binÃ¡rios
map_binario = {"Yes": 1, "No": 0}

# Aplicar a transformaÃ§Ã£o nas colunas desejadas
colunas_binarias = ['Churn', 'customer_Partner', 'customer_Dependents', 'phone_PhoneService', 'account_PaperlessBilling']

for col in colunas_binarias:
    dados[col] = dados[col].map(map_binario)
```

```python
#Exibir colunas binÃ¡rias
dados[colunas_binarias].head()
```

```python
#Padronizar e traduzir os nomes das colunas
dados = dados.rename(columns={
    'customer_gender': 'genero_cliente',
    'customer_SeniorCitizen': 'maior_65_anos',
    'customer_Partner': 'possui_parceiro',
    'customer_Dependents': 'dependentes',
    'customer_tenure': 'tempo_contrato_meses',
    'account_Charges_Monthly': 'cobranca_mensal',
    'account_Charges_Total': 'cobranca_total',
    'account_Contract': 'tipo_contrato',
    'phone_PhoneService': 'servico_telefonico',
    'phone_MultipleLines': 'mais_de_uma_linha_telefonica',
    'internet_InternetService': 'servico_internet',
    'internet_OnlineSecurity': 'servico_seguranca_online',
    'internet_OnlineBackup': 'servico_online_backup',
    'internet_DeviceProtection': 'servico_protecao_dispositivo',
    'internet_TechSupport': 'servico_suporte_tecnico',
    'internet_StreamingTV': 'servico_streaming_TV',
    'internet_StreamingMovies': 'servico_streaming_filmes',
    'account_PaperlessBilling': 'fatura_online',
    'account_PaymentMethod': 'forma_de_pagamento'

})
dados.head()
```

## 1. IntroduÃ§Ã£o

O objetivo deste relatÃ³rio Ã© analisar os dados de clientes da empresa de telecomunicaÃ§Ãµes Telecom X, identificando os principais fatores que levam Ã  evasÃ£o de clientes, conhecida como Churn.

### 2. ExtraÃ§Ã£o e PreparaÃ§Ã£o dos Dados

**a. Fonte dos dados:** Arquivo TelecomX_Data.json, disponibilizado pela Alura/Oracle.

**b. Local de armazenamento**: data/dadosAPI/TelecomX_Data.json.

**c. Processos realizados:**
* ExtraÃ§Ã£o do dataset via API.

## ConclusÃµes e Insights

Com base na anÃ¡lise exploratÃ³ria, podemos destacar os seguintes insights:

*   **Taxa de Churn:** Aproximadamente 26.5% dos clientes cancelaram seus serviÃ§os, representando uma taxa de evasÃ£o significativa.
*   **Tipo de Contrato:** Clientes com contratos mensais apresentam uma probabilidade muito maior de churn em comparaÃ§Ã£o com aqueles com contratos de um ou dois anos. Isso sugere que contratos de longo prazo promovem maior retenÃ§Ã£o.
*   **Tempo de Contrato (Tenure):** Clientes com menor tempo de contrato (principalmente nos primeiros 12 meses) tÃªm uma taxa de churn consideravelmente mais alta. A taxa de evasÃ£o diminui Ã  medida que o tempo de contrato aumenta.
*   **Idosos:** Clientes com mais de 65 anos demonstram uma taxa de churn maior do que clientes mais jovens.
*   **Dependentes:** Clientes sem dependentes tendem a ter uma taxa de churn maior do que aqueles com dependentes.
*   **MÃ©todo de Pagamento:** Clientes que utilizam o mÃ©todo de pagamento "Electronic check" parecem ter uma taxa de churn mais alta em comparaÃ§Ã£o com outros mÃ©todos.
*   **ServiÃ§os Contratados:** HÃ¡ uma tendÃªncia de que clientes com menos serviÃ§os contratados apresentem maior churn. Clientes com 0 ou 1 serviÃ§o tÃªm uma taxa de evasÃ£o notavelmente alta. Clientes com um maior nÃºmero de serviÃ§os tendem a permanecer mais tempo.
*   **Gastos:** Clientes que cancelam tendem a ter uma cobranÃ§a mensal mais alta e um gasto total acumulado menor, o que faz sentido dado o menor tempo de contrato.

## RecomendaÃ§Ãµes

Com base nas conclusÃµes, sugiro as seguintes recomendaÃ§Ãµes para reduzir a evasÃ£o de clientes:

*   **Programas de Fidelidade para Contratos Longos:** Oferecer incentivos e benefÃ­cios exclusivos para clientes que optam por contratos de um ou dois anos.
*   **EstratÃ©gias de RetenÃ§Ã£o para Novos Clientes:** Implementar aÃ§Ãµes de acompanhamento proativo e suporte dedicado para clientes nos primeiros meses de contrato, que sÃ£o mais propensos ao churn.
*   **Ofertas EspecÃ­ficas para Idosos:** Desenvolver planos e serviÃ§os adaptados Ã s necessidades de clientes com mais de 65 anos, com foco em simplicidade e suporte acessÃ­vel.

*   **AnÃ¡lise do MÃ©todo de Pagamento "Electronic Check":** Investigar as razÃµes pelas quais clientes que usam este mÃ©todo tÃªm maior churn. Pode haver problemas com o processo de pagamento, taxas ou insatisfaÃ§Ã£o geral.
*   **SegmentaÃ§Ã£o e Campanhas Direcionadas:** Utilizar os insights sobre os grupos de maior risco (contratos mensais, baixo tenure, idosos, sem dependentes, poucos serviÃ§os) para criar campanhas de marketing e retenÃ§Ã£o personalizadas.
